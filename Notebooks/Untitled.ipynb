```python
import pandas as pd
import numpy as np

# Semilla para reproducibilidad
np.random.seed(42)
```

```python
n_estudiantes = 500
```

```python
estudiante_id = np.arange(1, n_estudiantes + 1)

carrera = np.random.choice(
    ["Ingeniería de Sistemas", "Ingeniería Civil", "Administración", "Educación"],
    size=n_estudiantes
)

semestre = np.random.randint(1, 11, size=n_estudiantes)

horas_estudio = np.random.randint(5, 40, size=n_estudiantes)

carga_academica = np.random.randint(1, 6, size=n_estudiantes)

recursos = np.random.randint(1, 6, size=n_estudiantes)

apoyo_institucional = np.random.randint(1, 6, size=n_estudiantes)
```

```python
nivel_estres = (
    carga_academica * 0.4
    + (5 - recursos) * 0.3
    + (5 - apoyo_institucional) * 0.3
)

nivel_estres = np.clip(np.round(nivel_estres), 1, 5)
```

```python
promedio = (
    horas_estudio * 0.05
    + recursos * 0.8
    + apoyo_institucional * 0.6
    - nivel_estres * 0.7
    + np.random.normal(0, 1, n_estudiantes)
)

promedio = np.clip(np.round(promedio, 2), 0, 20)
```

```python
satisfaccion = (
    recursos * 0.4
    + apoyo_institucional * 0.4
    - nivel_estres * 0.2
)

satisfaccion = np.clip(np.round(satisfaccion), 1, 5)
```

```python
df = pd.DataFrame({
    "estudiante_id": estudiante_id,
    "carrera": carrera,
    "semestre": semestre,
    "horas_estudio": horas_estudio,
    "carga_academica": carga_academica,
    "recursos": recursos,
    "apoyo_institucional": apoyo_institucional,
    "nivel_estres": nivel_estres,
    "promedio": promedio,
    "satisfaccion": satisfaccion
})

df.head()
```

```python
ruta = "../data/sistema_educativo.csv"
df.to_csv(ruta, index=False)

print("Dataset generado y guardado correctamente ✅")
```
